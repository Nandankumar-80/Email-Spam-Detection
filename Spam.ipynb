```python
# 1. Import Libraries
import pandas as pd
import numpy as np
```

```python
# 2. Load Dataset
df = pd.read_csv("spam.csv", encoding='latin-1')
print(df.head(5))
```

```python
# 3. Data Cleaning
df = df[['v1', 'v2']]   # keep only needed columns
df.columns = ['label', 'message']

print(df.head())
```

```python
# 4. Convert Labels to Numeric
# ham = 0, spam = 1
df['label'] = df['label'].map({'ham': 0, 'spam': 1})
```

```python
# 5. Feature Extraction (Text → Numbers)
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
X = cv.fit_transform(df['message'])

y = df['label']
```

```python

# 6. Train-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# 7. Model Training (Naive Bayes)
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train, y_train)

print("Model Trained Successfully ✅")
```

```python

# 8. Prediction
y_pred = model.predict(X_test)
```

```python

# 9. Evaluation
from sklearn.metrics import accuracy_score, classification_report

print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nReport:\n", classification_report(y_test, y_pred))

# =========================
#  Test with Custom Message
# =========================

msg = ["Congratulations!  You have shortlisted for softer development in x company"]
msg_vec = cv.transform(msg)

prediction = model.predict(msg_vec)

if prediction[0] == 1:
    print("\nThis is SPAM ❌")
else:
    print("\nThis is NOT SPAM ✅")
```
